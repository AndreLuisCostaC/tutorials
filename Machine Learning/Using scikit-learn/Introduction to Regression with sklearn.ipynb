# The Boston House Price Database

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.manifold import TSNE, SpectralEmbedding, Isomap, MDS
```

```python
boston = load_boston()
X = boston.data
Y = boston.target

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, shuffle= True)
```

The dataset has instances $x_i \in {\rm I\!R}^{13}$, thus to visualize the dataset we will use an embedding technique to try and maintain the relative distance between instances whilst reducing the data to 2-dimensions. It can be seen with the following methods that different embedding strategies will lead to different distributions of the data. However, this can be used to view the feasibility of applying regression algorithms, if some structure exists embedded in the data then we should be confident that regression is possible on such a dataset.

```python
# Embed the features into 2 features using TSNE
X_embedded_iso  = Isomap(n_components=2).fit_transform(X)
X_embedded_mds  = MDS(n_components=2, max_iter=100, n_init=1).fit_transform(X)
X_embedded_tsne = TSNE(n_components=2).fit_transform(X)
X_embedded_spec = SpectralEmbedding(n_components=2).fit_transform(X)
```

```python
print('Description of the dataset: \n')

print('Input shape : ', X_train.shape)
print('Target shape: ', y_train.shape)

plt.plot(Y)
plt.title('Distribution of the prices of the homes in the Boston area')
plt.xlabel('Instance')
plt.ylabel('Price')
plt.show()

print('Embed the features into 2 features using Spectral Embedding: ', X_embedded_spec.shape)
print('Embed the features into 2 features using TSNE: ', X_embedded_tsne.shape)

fig = plt.figure(figsize=(12,5),facecolor='w')
plt.subplot(1, 2, 1)
plt.scatter(X_embedded_iso[:,0], X_embedded_iso[:,1], c = Y, cmap = 'hot')
plt.title('2D embedding using Isomap \n The color of the points is the price')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.scatter(X_embedded_mds[:,0], X_embedded_mds[:,1], c = Y, cmap = 'hot')
plt.title('2D embedding using MDS \n The color of the points is the price')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
plt.tight_layout()

fig = plt.figure(figsize=(12,5),facecolor='w')
plt.subplot(1, 2, 1)
plt.scatter(X_embedded_spec[:,0], X_embedded_spec[:,1], c = Y, cmap = 'hot')
plt.title('2D embedding using Spectral Embedding \n The color of the points is the price')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.scatter(X_embedded_tsne[:,0], X_embedded_tsne[:,1], c = Y, cmap = 'hot')
plt.title('2D embedding using TSNE \n The color of the points is the price')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
plt.tight_layout()
```

## Linear Regression

```python
lineReg = LinearRegression()
lineReg.fit(X_train, y_train)
print('Score: ', lineReg.score(X_test, y_test))

plt.plot(lineReg.predict(X_test))
plt.plot(y_test)
plt.show()
```

## Ridge Regression

```python
from sklearn import linear_model
reg = linear_model.Ridge (alpha = .5)
reg.fit(X_train, y_train)
print('Score: ', reg.score(X_test, y_test))

plt.plot(reg.predict(X_test))
plt.plot(y_test)
plt.show()
```

## Lasso Regression

```python
reg = linear_model.Lasso(alpha = 0.1)
reg.fit(X_train, y_train)
print('Score: ', reg.score(X_test, y_test))

plt.plot(reg.predict(X_test))
plt.plot(y_test)
plt.show()
```

## Polynomial fitting with a Perceptron

```python
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(interaction_only=True)
X_train_trans = poly.fit_transform(X_train).astype(int)
X_test_trans  = poly.transform(X_test).astype(int)

clf = Perceptron(fit_intercept=False, max_iter=10, tol=None, shuffle=False)
clf.fit(X_train_trans, y_train.astype(int))

print('Score: ', clf.score(X_test_trans, y_test.astype(int)))

plt.plot(clf.predict(X_test_trans))
plt.plot(y_test)
plt.show()
```

## Support Vector Regression

```python
from sklearn.svm import SVR
clf = SVR(C=1.1, epsilon=0.02)
clf.fit(X_train, y_train)
print('Score: ', clf.score(X_test, y_test))

plt.plot(clf.predict(X_test))
plt.plot(y_test)
plt.show()
```

Adding a polynomial kernel transformation and scaling the data appropriately. 

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

clf = SVR(C=0.1, epsilon=0.02, kernel='poly', degree = 5, max_iter = 1000)
clf.fit(scaler.transform(X_train), y_train)
print('Score: ', clf.score(scaler.transform(X_test), y_test))

plt.plot(clf.predict(scaler.transform(X_test)))
plt.plot(y_test)
plt.show()
```
