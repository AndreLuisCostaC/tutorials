```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 25
x = np.random.rand(n,3)
y = np.zeros((n,))
num_labels = 6
for i in range(num_labels):
    y[(i*n)//num_labels:((i+1)*n)//num_labels] = i
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
dists = np.asarray([np.linalg.norm(i-j) for i in x for j in x]).reshape(n,n)
```

```python
plt.imshow(dists)
plt.show()
```

```python
groups = []

for row in dists:
    labels = []
    points = []
    dis = []
    for i in np.argsort(row):
        if y[i] not in labels:
            labels.append(y[i])
            points.append(x[i])
            dis.append(row[i])
        else:
            dis.append(row[i])
            break
    groups.append([labels, points, dis])

[labels, points, dis]
```

```python
fig, ax = plt.subplots()

for i in groups:
    xx, yy = i[1][0][0], i[1][0][1]
    r = (i[2][-1] + i[2][-2])/2
    circle = plt.Circle((xx, yy), r, color='b', fill=True, alpha = 0.1)
    ax.add_artist(circle)

plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
radii = []

for row in dists:
    labels = []
    dis = []
    for i in np.argsort(row):
        if y[i] not in labels:
            labels.append(y[i])
            dis = row[i]
        else:
            dis = (dis + row[i])/2
            break
    radii.append(dis)

radii
```

```python
fig, ax = plt.subplots(figsize=(10,10))

for ix, i in enumerate(radii):
    circle = plt.Circle((x[ix, 0], x[ix, 1]), i, color='b', fill=False, alpha = 0.5)
    ax.add_artist(circle)

plt.scatter(x[:,0], x[:,1], c=y)
plt.xlim([-0.2,1.4])
plt.ylim([-0.2,1.4])
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from numpy import where

# Plots 2 features, with an output, shows the decision boundary
def plot3D(x, y):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    
    for ix, i in enumerate(radii):
        # draw sphere
        u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
        xx = i*np.cos(u)*np.sin(v) +x[ix, 0]
        yy = i*np.sin(u)*np.sin(v) +x[ix, 1]
        zz = i*np.cos(v) +x[ix, 2]
        ax.plot_wireframe(xx, yy, zz, color="r", alpha = 0.1)
    
    ax.scatter(x[:, 0], x[:, 1],x[:, 2], marker='o', c=y)
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')

    axes = plt.axis()
    plt.show()
    
plot3D(x, y)
```
