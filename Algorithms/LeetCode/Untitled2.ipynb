```python

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
house_price = [245, 312, 279, 308, 199, 219, 405, 324, 319, 255]
size = [1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]
size2 = np.array(size).reshape((-1, 1))
#fitting into the model
regr = linear_model.LinearRegression()
regr.fit(size2, house_price)
print("Coefficients: \n", regr.coef_)
print("intercept: \n", regr.intercept_)
#############################
#formula obtained for the trained model
def graph(formula, x_range):
   x = np.array(x_range)
   y = eval(formula)
   plt.plot(x, y)
#plotting the prediction line 
graph('regr.coef_*x + regr.intercept_', range(1000, 2700))
print(regr.score(size2, house_price))
plt.scatter (size,house_price, color='black')
plt.ylabel('house price')
plt.xlabel('size of house')
plt.show()
```

```python
def myAtoi(str):
    """
    :type str: str
    :rtype: int
    """
    first_num_found = 0
    found_sign = 0
    
    res_string = ""
    for i in str:
        
        if first_num_found == 0 and (i == '-' or i == '+'):
            res_string += i
            found_sign = 1
        elif i == ' ' and first_num_found == 0 and found_sign == 0:
            pass
        elif ord(i) >= 48 and ord(i) <= 57:
            res_string += i
            first_num_found = 1
        else:
            break
        
        print(i, '   ', i.encode('utf-8'))
    
    if len(res_string) == 1 and (res_string[0] == '-' or res_string[0] == '+'):
        return 0
    
    if len(res_string) > 1 and (res_string[1] == '+' or res_string[1] == '-'):
        return 0
    
    print(res_string)
    print(type(res_string))
    print(int(res_string))
    
    if len(res_string) > 11 and res_string[0] == '-':
        return -2147483648
    elif len(res_string) == 11 and int('-2147483648') >= int(res_string):
        return -2147483648
    
    if len(res_string) == 11 and res_string[0] == '+' and int('2147483647') >= int(res_string):
        pass
    elif len(res_string) > 10 and res_string[0] != '-':
        return 2147483647
    elif len(res_string) == 10 and int('2147483647') <= int(res_string):   
        return 2147483647
    
    if len(res_string) == 0:
        return 0
    else: return int(res_string)
```

```python
def findContentChildren(g, s):
    """
    :type g: List[int]
    :type s: List[int]
    :rtype: int
    """
    g = sorted(g)
    s = sorted(s)
    count = 0
    for child in g:
        for ix, cookie in enumerate(s):
            if child <= cookie:
                del s[ix]
                count += 1
                break

    return count
```

```python
def reverse(x):
    """
    :type x: int
    :rtype: int
    """
    # Convert to a string for easier manipulation
    x = str(x)
    
    # Check for a sign
    sign = ''
    if x[0] == '-':
        sign = x[0]
        x = x[1:len(x)]
    
    # Reverse the string
    num = x[::-1]
    
    start = 0
    for ix, i in enumerate(num):
        if i != 0:
            start = ix
            break
        
    num = int(sign + num[start:len(num)])
    
    if num > 2147483647 or num < -2147483647: return 0
    return num
```

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        
        l = nums1 + nums2
        l = sorted(l)
        
        length = len(l)
        if length % 2 == 0:
            val = (l[(length-1)//2]+l[(length+1)//2])/2
        else:
            val = l[length//2]
        return val
```

```python
class Solution:
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        if numRows == 0:
            return []
        
        triangle = [[1]]
        
        for i in range(numRows-1):
            temp = []
            for ix, j in enumerate(triangle[i]):
                if ix == 0:
                    temp.append(1)
                else:
                    temp.append(triangle[i][ix-1]+triangle[i][ix])
            temp.append(1)
            triangle.append(temp)
        return triangle
```

```python
from copy import deepcopy
import numpy as np
import time

class Solution:
    
    def cleanList(self, wordList):
        new_wordList = []
        for word1 in wordList:
            valid = 0
            for word2 in wordList:
                if self.string1off(word1, word2): valid = 1
            if valid == 1: new_wordList.append(word1)
        return new_wordList
    
    def string1off(self, str1, str2):
        diff = 0
        for i in range(len(str1)):
            if str1[i] != str2[i]: diff += 1
        if diff == 1: return True
        else: return False
    
    def findLadders(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: List[List[str]]
        """
        print(len(wordList))
        wordList = self.cleanList(wordList)
        print(len(wordList))
        
        if beginWord not in wordList:
            wordList = [beginWord] + wordList
        
        verbose = 0
        
        t0 = time.time()
        
        one_off = np.zeros((len(wordList), len(wordList)))
        for i, word1 in enumerate(wordList):
            for j, word2 in enumerate(wordList):
                if i>j and self.string1off(word1, word2): one_off[i,j] = 1
        one_off = one_off + one_off.T
        print(one_off)
        
        print('Get dists', time.time()-t0)
        
        results = [[beginWord]]
        
        stop = 0
        count = 0
        while count <= len(wordList):
            count += 1
            
            temp = []
            for result in results:
                last_word = result[-1]
                
                temp_wordList = [i for i in wordList if i not in result]
                
                ix = wordList.index(last_word)
                conn_nodes = np.where(one_off[ix, :] == 1)
                
                words = []
                for i in conn_nodes[0]:
                    words.append(wordList[i])
        
                for word in words:
                    if word not in result:
                        temp.append(result + [word])
                        if word == endWord:
                            stop = 1
                            
                if verbose == 1:   
                    print('-----------------------------------------------------------')
                    print(last_word)
                    print(ix)
                    print('connections: ', conn_nodes[0], '    length:', len(conn_nodes[0]))
                    print(words)
            if verbose == 1: print(temp)
            
            # No possible solutions
            if len(temp) == 0:
                return []
            
            results = deepcopy(temp)

            if stop == 1:
                print('STOP!!!')
                break
        
        temp = []
        for result in results:
            if result[-1] == endWord:
                temp.append(result)
                
        return temp
```
