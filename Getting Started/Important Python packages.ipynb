### Plotting multiple lines

```python
n = 4
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.random.rand(n)
y4 = 2*np.random.rand(n)
y5 = 2*np.random.rand(n)
y6 = 2*np.random.rand(n)
y = [y1, y2, y3, y4, y5, y6]
x = range(n)

colors = ['#1B9E77', '#D95F02', '#7570B3', '#E7298A', '#66A61E', '#E6AB02']
colors = ['#E41A1C', '#377EB8', '#4DAF4A', '#984EA3', '#FF7F00', '#1B9E77']
markers = ['.', 'x', ',', 'o', 'v', '^']

plt.rcParams.update({'font.size': 14})
plt.figure(num=2, figsize=(16, 4), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(1,1,1)
lines = []
for ix, i in enumerate(y):
    line, = plt.plot(x, i, linewidth = 2.5, color=colors[ix], marker = markers[ix], label = 'y'+str(ix))
    lines.append(line)
    
plt.legend(handles=lines, bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=6, mode="expand", borderaxespad=0.)
plt.rcParams.update({'font.size': 14})
plt.xlabel('The x-axis')
plt.ylabel('The y-axis')
plt.tight_layout()
plt.show()
```

## 3D Plot

```python
import matplotlib.cm as cm

n = 100
x = np.random.randn(int(n/2), 2)+1
x = np.append(x, np.random.randn(int(n/2), 2)+3, axis=0)
y = np.zeros(int(n/2),)
y = np.append(y, np.ones(int(n/2),))

plt.scatter(x[:,0], x[:,1], c = y, cmap='autumn')
plt.colorbar()
plt.title('A plot')
plt.show()
```

```python
import time
import pylab as pl
from IPython import display

n = 1280
r = 4

a = np.zeros((n, n))
nx, ny = (n, n)
x = np.linspace(-r, r, nx)
y = np.linspace(-r, r, ny)
xx, yy = np.meshgrid(x, y, sparse=True)

a = np.exp(-1*(xx**2 + yy**2))
a += np.sin(2 * np.pi * (xx**2 + yy**2))
a += np.sqrt(xx**2 + yy**2)
a += np.sqrt(xx**2+yy**2)*np.cos(50/(xx**2+yy**2))
a += np.cos(100/(xx**2+yy**2))
a += np.sin(20/(xx**2+yy**2))

plt.figure(num=1, figsize=(4, 4))
ranges = [0.001, 0.005, 0.01, 0.05, 0.1, 0.15, 0.2, 0.5, 1, 2, 3, 4, 5]
screens = []
for r in ranges:
    a = np.zeros((n, n))
    nx, ny = (n, n)
    x = np.linspace(-r, r, nx)
    y = np.linspace(-r, r, ny)
    xx, yy = np.meshgrid(x, y, sparse=True)

    a = np.exp(-1*(xx**2 + yy**2))
    a += np.sin(2 * np.pi * (xx**2 + yy**2))
    a += np.sqrt(xx**2 + yy**2)
    a += np.sqrt(xx**2+yy**2)*np.cos(50/(xx**2+yy**2))
    a += np.cos(100/(xx**2+yy**2))
    a += np.sin(20/(xx**2+yy**2))
    
    screens.append(a)
    
for i in range(len(ranges)):
    plt.imshow(screens[i], cmap='afmhot', interpolation='nearest')
    display.clear_output(wait=True)
    display.display(pl.gcf())
    time.sleep(0.02)
display.clear_output(wait=True)
```
